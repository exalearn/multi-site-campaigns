```python
%matplotlib inline
import re
import numpy as np
import pandas as pd

pd.options.display.max_columns = None

from IPython.display import display, HTML
from pathlib import Path
from matplotlib import pyplot as plt
from datetime import datetime

import matplotlib.pylab as pylab
params = {
    'axes.labelsize': 10,
    'xtick.labelsize': 7,
    'ytick.labelsize': 7
}
pylab.rcParams.update(params)
```

```python
# ALIAS
funcx = 'FuncX'
proxystore = 'Proxify'
colmena = 'Hive'
globus = 'Globus'
```

```python
def extract(string, key, dtype):
    if dtype is bool:
        return True if re.search(' {}=((True)|(False)) '.format(key), string).group(1) == 'True' else False
    if dtype is float:
        return float(re.search(' {}=(\d+(\.\d+)?(?:[eE][+\-]?\d+)?)'.format(key), string).group(1))
    if dtype is int:
        return float(re.search(' {}=(\d+) '.format(key), string).group(1))

def get_result_metrics(filepath):
    # get lines with results
    lines = []
    marker = 'Got result: '
    with open(filepath) as f:
        for line in f.readlines():
            if marker in line:
                lines.append(line.split(marker, 1)[1])
    
    # extract metrics
    runs = []
    for run in lines:
        if not extract(run, 'success', bool):
            continue
        data = {
            'time_created': extract(run, 'time_created', float),
            'time_input_received': extract(run, 'time_input_received', float),
            'time_compute_started': extract(run, 'time_compute_started', float),
            'time_compute_ended': extract(run, 'time_compute_ended', float),
            'time_result_sent': extract(run, 'time_result_sent', float),
            'time_result_received': extract(run, 'time_result_received', float),
            'time_running': extract(run, 'time_running', float),
            'time_serialize_inputs': extract(run, 'time_serialize_inputs', float),
            'time_deserialize_inputs': extract(run, 'time_deserialize_inputs', float),
            'time_serialize_results': extract(run, 'time_serialize_results', float),
            'time_deserialize_results': extract(run, 'time_deserialize_results', float),
            'time_result_received': extract(run, 'time_result_received', float),
            'time_async_resolve_proxies': extract(run, 'time_async_resolve_proxies', float),
            'time_task_submitted': extract(run, 'time_task_submitted', float),
            'time_task_received': extract(run, 'time_task_received', float),
        }
        

        data['Time on Worker'] = (data['time_compute_ended'] + data['time_serialize_results']) - data['time_compute_started'] 
        data['Thinker->Task Server'] = data['time_input_received'] - (
                data['time_created'] + data['time_serialize_inputs'])
        data['Task Server->Worker'] = data['time_compute_started'] - data['time_task_submitted']
        #data['Worker->Task Server'] = data['time_task_received'] - (data['time_compute_ended'] + data['time_serialize_results'])
        data['Worker->Task Server'] = data['time_result_sent'] - (
                data['time_compute_started'] + data['time_running'] + 
                data['time_deserialize_inputs'] + data['time_serialize_results'] +
                data['time_async_resolve_proxies'])
        data['Task Server->Thinker'] = data['time_result_received'] - (
                data['time_result_sent'] + data['time_deserialize_results'])
        data['Serialization'] = sum(data[key] for key in data if 'serialize' in key)
        data['Task Lifetime'] = data['time_result_received'] - data['time_created']
        
        runs.append(data)
    
    return pd.DataFrame(runs)

def aggregate_runs(run_paths, mean=False, stddev=False, drop=False):
    data = []
    for path in run_paths:
        results = pd.Series({'path': path})

        with open(path) as f:
            first_line = f.readline()
            timestamp = first_line.split(' - ')[0]
            proxystore_type = re.search('ps_backend=([^\s]+),', first_line).group(1)
            if proxystore_type == 'None':
                results['proxystore'] = 'none'
            elif proxystore_type == "'file'":
                results['proxystore'] = 'file'
            elif proxystore_type == "'redis'":
                results['proxystore'] = 'redis'
            elif proxystore_type == "'globus'":
                results['proxystore'] = 'globus'
            else:
                results['proxystore'] = 'error'
            results['reuse_data'] = True if re.search('reuse_data=((True)|(False))', first_line).group(1) == 'True' else False
            results['task_count'] = int(re.search('task_count=(\d+),', first_line).group(1))
            results['task_input_size'] = float(re.search('task_input_size=(\d+(\.\d+)?(?:[eE][+\-]?\d+)?),', first_line).group(1))
            results['task_interval'] = float(re.search('task_interval=(\d+(\.\d+)?(?:[eE][+\-]?\d+)?),', first_line).group(1))
            results['task_output_size'] = float(re.search('task_output_size=(\d+(\.\d+)?(?:[eE][+\-]?\d+)?),', first_line).group(1))
            results['time_start'] = datetime.strptime(timestamp, "%Y-%m-%d %H:%M:%S,%f").timestamp()

        function_results = get_result_metrics(path)
        results['n_tasks'] = len(function_results.index)
        results['total'] = None
        if drop:
            function_results.drop(function_results.tail(1).index,inplace=True)
            function_results.drop(function_results.head(5).index,inplace=True)
        if mean:
            results = pd.concat([results, function_results.mean()])
        else:
            results = pd.concat([results, function_results.median()])
            
        if stddev:
            fstd = function_results.std().add_suffix('_std')
            results = pd.concat([results, fstd])
            
                
        data.append(results)
        
    return pd.concat(data, axis=1).T
```

```python
cols = [
    'Thinker->Task Server',
    'Task Server->Thinker',
    'Serialization',
    'time_async_resolve_proxies',
    'time_running',
    'total'
]
ind = np.arange(len(cols))

rundir = 'runs_final/50T-0D-1I-0O-1N'
run_paths = list(map(str, Path(rundir).rglob('*runtime.log')))
results = aggregate_runs(run_paths)
results['total'] = results[cols].sum(axis=1)
display(results)

none_unique = results.loc[results['proxystore'] == 'none'].loc[results['reuse_data'] == False][cols]
file_unique = results.loc[results['proxystore'] == 'file'].loc[results['reuse_data'] == False][cols]
redis_unique = results.loc[results['proxystore'] == 'redis'].loc[results['reuse_data'] == False][cols]

width = 0.25
fig, ax = plt.subplots(1, 1, figsize=(8,4), dpi= 100)

ax.bar(ind, none_unique.to_numpy()[0], width, label='Default')
ax.bar(ind+width, file_unique.to_numpy()[0], width, label='ProxyStore(File)')
ax.bar(ind+2*width, redis_unique.to_numpy()[0], width, label='ProxyStore(Redis)')


fig.suptitle('1 Node, {} tasks, {}MB input, {}s interval'.format(results['n_tasks'].iloc[0], results['task_input_size'].iloc[0], results['task_interval'].iloc[0]))
plt.setp(ax, ylabel='Median Time per Task (s)', xticks=ind + width/2, xticklabels=cols)
plt.setp(ax.get_xticklabels(), rotation=35, ha='right')
ax.set_title('Unique task input')
ax.legend(loc='best')
fig.tight_layout()
plt.show()
```
